```python
import pandas as pd

# Step 1: Load CSV into pandas DataFrame
# Replace 'your_file.csv' with your actual file path
df = pd.read_csv('bank-additional.csv', sep=';')

pd.set_option('display.max_columns', None)
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Step 2: Select the two columns you want to visualize
# Replace 'column_x' and 'column_y' with actual column names
x = df['column_x']
y = df['column_y']

# Step 3: Plot using matplotlib
plt.figure()
plt.plot(x, y)

# Step 4: Labeling (recommended for clarity)
plt.xlabel('column_x')
plt.ylabel('column_y')
plt.title('column_y vs column_x')

# Step 5: Display the plot
plt.show()
```
